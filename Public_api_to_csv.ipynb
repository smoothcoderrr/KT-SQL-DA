```python

import pandas as pd
import requests
from google.colab import files
```

```python
url="https://api.coincap.io/v2/assets"
header={"Content-Type":"application/json","Accept-Encoding":"deflate"}
response=requests.get(url,headers=header)
print(response)
```

```python
response_data=response.json()
print(response_data)
```

```python
data=pd.json_normalize(response_data,'data')
data.head()
```

```python
data.to_csv("STOCK_API.csv",index=False)
files.download('STOCK_API.csv')
```
